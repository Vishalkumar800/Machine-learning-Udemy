# Multiple Linear Regression Project

## When Do We Use Multiple Linear Regression?

    Multiple Linear Regression is used when we want to predict a continuous numerical value using two or more independent variables.
    For example, in house price prediction, the price of a house depends on several factors such as:

    House size
    Number of bedrooms
    Age of the house
    Distance from the city

## The model finds a best-fit hyperplane that minimizes the error and gives us the predicted value.

    Common Examples:

    House price prediction
    Salary prediction
    Sales forecasting
    Temperature prediction

    When Should We NOT Use Multiple Linear Regression?
    We should not use Multiple Linear Regression when the target variable is categorical, such as:

        Yes / No
        Pass / Fail
        Buy / Not Buy

```python
import pandas as pd

data = pd.read_csv("house_data_large.csv")

x = data.iloc[: , :-1].values
y = data.iloc[: , -1].values

data
```

```python
import pandas as pd

poly_data = pd.read_csv("poly_data.csv")
poly_data.head()
```

```python
x = poly_data.iloc[: , :-1]
y = poly_data.iloc[: , -1]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=90)
```

```python
poly_data["Quality"].unique() , poly_data["Location"].unique()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder , OneHotEncoder

categories = [["Low","Medium","High","Premium"]]

ct = ColumnTransformer(
    transformers= [
        ("ord" , OrdinalEncoder(categories=categories) , ["Quality"]),
        ("ohe" , OneHotEncoder(drop= "first" , sparse_output=False), ["Location"])
    ],
    remainder="passthrough"
)

ct.fit(x_train)

x_train_final = ct.transform(x_train)
x_test_final = ct.transform(x_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree= 2)

pf.fit(x_train_final)

x_train_poly = pf.transform(x_train_final)
x_test_poly = pf.transform(x_test_final)
```

```python
from sklearn.linear_model import LinearRegression

lg = LinearRegression()

lg.fit(x_train_poly , y_train)
```

```python
lg.score(x_test_poly , y_test)
```
