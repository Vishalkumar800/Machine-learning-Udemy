# Seaborn
      
      Seaborn is a Python data visualization library built on top of Matplotlib that provides a high-level interface for creating attractive and informative statistical graphics.

      Behind the scenes, seaborn uses matplotlib to draw its plots.
      Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.

  ### This means that Seaborn can take an entire dataset (like a Pandas DataFrame or array) as input and automatically:

         Understand the meaning of the data (semantic mapping)

         For example:

          Which column is x-axis

          Which column is y-axis

          Which column represents categories (colors, groups, etc.)


## Difference
      
      Matplotlib is a low-level plotting library that provides extensive control over every aspect of a plot, allowing users to customize figures in detail. However, this flexibility often requires more code and effort to achieve visually appealing results.

      whereas,
      Seaborn is a high-level visualization library built on top of Matplotlib that simplifies the process of creating attractive and informative graphics with less code.


      Another major difference lies in their default appearance. Matplotlib offers basic styling by default, and users typically need to manually adjust styles to improve the visual quality of plots.

      Seaborn, on the other hand, provides aesthetically pleasing themes and color palettes automatically, making visualizations more professional without additional configuration.

## Important Line 
     
     Matplotlib usually requires users to extract and pass data manually, whereas Seaborn integrates seamlessly with Pandas DataFrames, allowing users to directly reference column names when creating plots. This makes Seaborn more convenient for data analysis workflows.

### Matplotlib
      
      plt.plot(df["age"], df["salary"])

      you are manually extracting the columns from the DataFrame.

### Seaborn
      
      sns.scatterplot(data=df, x="age", y="salary")
      
      Here, you simply provide the DataFrame and column names, and Seaborn retrieves the data itself.


```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

# Scatter plot

### hue 
      the hue parameter is used to add a categorical separation to your plots by coloring different groups differently.
      It’s available in many Seaborn functions like histplot, scatterplot, barplot, etc.

      You pass a categorical variable to hue.
      Seaborn automatically assigns different colors to each category.
      A legend is added to indicate which color corresponds to which category.


### style
      
      The style parameter in Seaborn is used to differentiate data points by assigning different marker shapes or line patterns based on a variable.

### palette

       The palette parameter in Seaborn is used to control the colors that are applied to different categories in a plot, especially when you are using the hue parameter.

### In simple terms:
       palette decides which set of colors (color theme) will be used in the visualization.

```python
sns.scatterplot(x = "total_bill" , y ="tip" , data=tips , hue="sex" , style="smoker" , palette="viridis")
```
