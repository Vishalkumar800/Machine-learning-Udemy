# Types of Machine Learning

1. Supervised Learning
2 . Unsupervised Learning
3 . Reinforcement Learning


Supervised :  
           SuperVised Learning models can make after seeing a lot data with their correct answers and then discovering the connections between the elements in the data that produce correct answer.


Two types of the most common uses for supervised learning are :
  1. Regression
  2. Classification

  Regression : A regression model predict a numeric value.

  Classification :
          A classification model predict a category.
        


# Unsupervised Learning :
  
  Unsupervised Learning , model make predictions by being given data that does not contain any correct answer. model goal is to indentify meaningful 
  pattern in the data.

  Commonly unsupervised learning model employs a technqiue called clustering.

# Reinforcement Learning

In Reinforcement leanring, model make predicitions by getting or penalties based on actions performed within the environment.

# Simple Linear Regression . 

Simple linear regression is a machine learning method in which we predict one value using one input value by drawing the best fit line through the data.

Note :
In simple linear Regression only one feature column and label are column are present.


```python
import pandas as pd

data = pd.read_csv("dataset.csv")
x = data.iloc[ : ,:-1].values # Feature
y = data.iloc[: , 1].values # Label
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.3 , random_state= 20 )
# random state is used to get same train test split every time when we run the code.
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train , y_train)
```

```python
x_test , y_test
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_test,y_test , color = 'red')
plt.plot(x_test , y_pred )
plt.xlabel("Student Study hours")
plt.ylabel("Student Scores")
plt.show()
```

```python
model.coef_ , model.intercept_
```
