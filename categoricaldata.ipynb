```python
import pandas as pd

data = pd.read_csv("data_preprocessing_dataset.csv")
x = data.iloc[: , :-1]
y = data.iloc[: , -1]


data.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain ,ytest = train_test_split(x , y , test_size= 0.2 , random_state= 48)
```

# Handling Categorical data

1. Ordinal Categorical data
1 . Nominal categorical data

Ordinal Categorical data:

     if there is any order or rank provide in data(column) then we say that this is ordinal categorical data.
     Free < basic < Premiun  < Pro

# ------ Ordinal encoding

Nominal Categorical data:

     if there is no order or rank provide in column then we say that is is nominal categorical data.
# ------ Nominal Encoding

# Note :
You mainly apply ordinal encoding and nominal encoding on features (x) , not labels.
 

```python
xtrain["Satisfaction"].unique() 
```

```python
from sklearn.preprocessing import OrdinalEncoder
#  Ordinal Encoding, the first element in the list is assigned the lowest value, and the last element is assigned the highest value.

categories = [["Free", "Basic" ,"Premium" , "Pro"],  ["School" , "UG" ,"PG" ,"Diploma"] , ["VeryLow" ,"Low", "Medium", "High" ,"VeryHigh"]]

oe = OrdinalEncoder(categories=categories)

oe.fit(xtrain[["Subscription", "EducationLevel" ,"Satisfaction"]])

c = oe.transform(xtrain[["Subscription" , "EducationLevel" ,"Satisfaction"]])
d = oe.transform(xtest[["Subscription" , "EducationLevel" ,"Satisfaction"]])

c
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output= False , drop = "first")
listcolumnx = ["Gender","City", "Device","PaymentMode","PurchaseCategory"]
ohe.fit(xtrain[listcolumnx])

ohe.transform(xtrain[listcolumnx])
ohe.transform(xtest[listcolumnx])
```

```python
from sklearn.preprocessing import LabelEncoder

lb  = LabelEncoder()

lb.fit(ytrain)


y_train_final = lb.transform(ytrain)
y_test_final = lb.transform(ytest)
```

```python
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers= [
    ("ord" , OrdinalEncoder(categories= categories) , ["Subscription", "EducationLevel" ,"Satisfaction"]) ,
    ("OneHotEncoder" , OneHotEncoder(sparse_output= False , drop= "first") , ["Gender","City", "Device","PaymentMode","PurchaseCategory"] )
], remainder= "passthrough")

# drop  =  means: remove the columns on which no transformation is applied.
# passthrough  =  means: keep the columns on which no transformation is applied.
```

```python
ct.fit(xtrain)
x_train_final = ct.transform(xtrain)
xtest_final = ct.transform(xtest)
```

ColumnTransformer :The ColumnTransformer is a powerful tool in sklearn for applying different preprocessing transformations to specific columns within a dataset. This allows processing based on the nature of each feature.

transformers : 
("name", transformer, columns)
