# Seaborn

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
tips
```

```python
sns.scatterplot(data=tips , x = "total_bill" , y= "tip" , hue="sex",style="time") # it is axis level function
```

```python
# figure level function

sns.relplot(data= tips , x = "total_bill" , y= "tip" , kind="scatter") # line
```

# Line graph

```python
data = pd.read_csv("dataset.csv")
sns.lineplot(data=data , x = "StudyHours" , y = "Marks")
```

# facet plot

```python
sns.relplot(data=tips, x="total_bill" , y="tip" , kind="scatter", col= "sex")
```

```python
# figure level function -> displot
# axis level function -> histplot , kdeplot , ecdfplot , rugplot

sns.histplot(data=tips , x="total_bill")
```

```python
sns.displot(data=tips , x = "total_bill" , kind="hist")
```

```python
sns.displot(data=tips , x="total_bill" , kind="hist" , hue="sex")
```

```python
sns.displot(data=tips , x="total_bill", col="sex" , kind="hist", element ="step")
```

# KDE plot

```python
sns.kdeplot(data=tips , x= "total_bill")
```

```python
sns.kdeplot(data= tips , x="total_bill")
sns.rugplot(data=tips , x="total_bill")
```
