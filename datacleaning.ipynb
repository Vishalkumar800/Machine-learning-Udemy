# backward filling and forward filling

  ## bfill() fills empty values with the value that comes after them.
        
        If a value is missing, it takes the next value (below it) and puts it there.
        Simple :
               Look down and copy the next value.

  ## ffill() fills empty values with the value that comes before them.
        
        If a value is missing, it takes the previous value (above it) and puts it there.
        Simple word:
                Look up and copy the previous value.


        axis=0 → column-wise (top to bottom)

        axis=1 → row-wise (left to right)

```python
data.bfill()
```


   ## Mode 
       Mode is the value that appears the most. 
       Numbers:
            2, 4, 4, 5, 6

       Here 4 appears two times (more than others).
       So,

           Mode = 4

    
  ## How to find mode  in column 

        data["ColumnName"].mode() then put in fillna()


```python
data["Sex"].fillna(data["Sex"].mode()[0], inplace= True)
```

# How to select a specific type of column.

 ## To get a specific type of column use :

     data.select_dtypes(include= "object")

 ## To get columns name of specific type of data use :
     
     data.select_dtypes(include = "object").columns



```python
data.select_dtypes(include= "object").isnull().sum()
```

# How to apply mode on mutliple columns 

```python
for i in data.select_dtypes(include="object").columns:
    data[i].fillna(data[i].mode()[0])
```

# Outlier 

     Outliers are data points that significantly deviate from the overall pattern of a dataset.

     
     1. They can appear as unusually high or low values compared to the rest of the data. 
     2. Can affect mean, variance, and model performance.
     3. Detected using statistical and visual methods.
     4. Important to analyze before removing or treating them

# Why Outlier Occur 
    
  ## Data Entry Errors:
       When we enter data manually, typing mistakes can happen.
       For example, writing age 20000 instead of 20. This creates a wrong value.

  ## Measurement Errors
       If a machine or measuring instrument is not working properly, it can give incorrect readings.
       For example, a faulty thermometer may show very high or very low temperature.

  ## Intentional Outliers
       Sometimes people change data on purpose.
       This happens in cases like fraud or data manipulation.
       
  ## Data Processing Errors
       Mistakes can happen while collecting, cleaning, or transforming data.
       For example, applying the wrong formula or mixing up columns.
