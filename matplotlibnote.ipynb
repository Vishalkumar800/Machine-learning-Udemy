
# 📊 Bar Width Range in Matplotlib

In Matplotlib, the width parameter in plt.bar():

Accepts a float value

The default width is 0.8

The typical range of width is between 0 and 1

Explanation:

width > 0 → bar is visible

width < 1 → bars do not overlap

width = 1 → bars touch each other

Very small values (e.g. 0.2) → very thin bars

In Matplotlib, we can assign different colors to each bar by passing a list of colors to the color parameter. Each color in the list corresponds to a bar in the bar chart.

## width = 0.5 → thinner bars

## width = 0.8 → default bar width


# Scatter Plot
A scatter plot is a graph where we plot data in the form of points (dots) to show the relationship between two values (X and Y).

```python
import matplotlib.pyplot as plt

months = ['jan','feb','march','april','may']
sales = [200,300,50,400,250]

plt.scatter(months , sales ,label =" Company Sales Data")
plt.xlabel("months")
plt.ylabel("Sales in Units")
plt.legend()
plt.grid(True)
plt.show()
```

```python
months = ['jan','feb','march','april','may']
sales = [200,300,50,400,250]

fig , ax = plt.subplots()

ax.plot(months , sales , marker = "*" , color = "purple")
ax.set_title("Sales Data")
ax.set_xlabel("Month")
ax.set_ylabel("Sales Data")
plt.show()
```

```python
import pandas as pd

data = pd.read_csv("multiple_linear_data.csv")
x = data.iloc[: ,:-1].values
y = data.iloc[: , -1].values
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=90)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()

lg.fit(x_train, y_train)
```

```python
lg.predict(x_test)
```

```python
lg.score(x_test, y_test)
```

```python
from mpl_toolkits.mplot3d  import Axes3D
import numpy as np

x1 = x_train[: , 0]
x2 = x_train[: , 1]

y_actual = y_train

y_prediction = lg.predict(x_test)

fig  = plt.figure()
ax = fig.add_subplot(111, projection ="3d")

ax.scatter(x1 , x2 , y_actual, color ="blue")

x1_range = np.linspace(x1.min() , x1.max() , 50)
x2_range = np.linspace(x2.min() , x2.max() , 50)

x1_grid , x2_grid = np.meshgrid(x1_range , x1_range)

y_grid = lg.intercept_ + lg.coef_[0] * x1_grid + lg.coef_[1] * x2_grid

ax.plot_surface(x1_grid , x2_grid , y_grid )

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")


plt.show()
```

# Subplot

By using subplot you can draw multiple plots in one figure.

## The subplot() function takes three arguments that describes the layout of the figure.

    The layout is organized in rows and columns, which are represented by the first and second argument.

    The third argument represents the index of the current plot.
